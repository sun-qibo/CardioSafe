```python
import pandas as pd

df_de = pd.read_excel("GSE123976_HF.v.NF_DESeq2.xlsx")
display(df_de)
```

```python
df_de['ensembl_gene_id'] = df_de['ensembl_gene_id'].str.split('.').str[0]
df_de.set_index('ensembl_gene_id', inplace=True)
sum(df_de['pvalue']<0.05)
```

```python
df_de_filtered = df_de[df_de['pvalue']<0.05]
df_de_filtered["absolute_log2FoldChange"] =  df_de_filtered["log2FoldChange"].abs()
df_de_filtered_top500 = df_de_filtered.sort_values(by="absolute_log2FoldChange", ascending=False).head(500)
df_de_filtered_top500 = df_de_filtered_top500.reset_index().rename(columns={'ensembl_gene_id': 'gene'})
df_de_filtered_top500 = df_de_filtered_top500[['gene', 'log2FoldChange']]
display(df_de_filtered_top500)
```

```python

df_de_filtered_top500.to_csv("top500_genes.txt", index=False, header=True)
```
